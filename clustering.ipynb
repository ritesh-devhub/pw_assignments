### Q21

```
# Q21
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
X,_=make_blobs(n_samples=300,centers=4)
k=KMeans(n_clusters=4).fit(X)
plt.scatter(X[:,0],X[:,1],c=k.labels_)
plt.show()
```

**Answer:** Output shows clustering result.

### Q22

```
# Q22
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
X=load_iris().data
labels=AgglomerativeClustering(n_clusters=3).fit_predict(X)
print(labels[:10])
```

**Answer:** Output shows clustering result.

### Q27

```
# Q27
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
X=PCA(2).fit_transform(load_digits().data)
k=KMeans(n_clusters=10).fit(X)
plt.scatter(X[:,0],X[:,1],c=k.labels_)
plt.show()
```

**Answer:** Output shows clustering result.

### Q28

```
# Q28
from sklearn.metrics import silhouette_score
for k in range(2,6):
    model=KMeans(n_clusters=k).fit(X)
    print(k,silhouette_score(X,model.labels_))
```

**Answer:** Output shows clustering result.

### Q29

```
# Q29
from scipy.cluster.hierarchy import dendrogram, linkage
Z=linkage(X,'average')
dendrogram(Z)
plt.show()
```

**Answer:** Output shows clustering result.
